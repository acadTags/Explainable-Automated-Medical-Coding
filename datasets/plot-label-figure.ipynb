```python
import matplotlib.pyplot as plt
import matplotlib.axes as axes
from collections import defaultdict
import numpy as np
```

```python
def output_list_freq_sorted(filename):
    with open(filename,encoding='utf-8') as f_content:
        content = f_content.readlines()
    content = [x.strip() for x in content]
    # create dictionary, label - label frequency
    dict_label_freq = defaultdict(int)
    for line in content:
        labelset = line.split("__label__")[1]
        for label in labelset.split():
            dict_label_freq[label] += 1
    #print(len(dict_label_freq),dict_label_freq)
    list_label_freq_sorted = sorted(dict_label_freq, key=dict_label_freq.get, reverse=True)
    #print(list_label_freq_sorted)
    list_freq_sorted = [dict_label_freq[ele] for ele in list_label_freq_sorted]
    #print(list_freq_sorted)
    return list_freq_sorted
```

```python
list_freq_sorted = output_list_freq_sorted(r'..\datasets\mimiciii_train_full_th_50_covid_shielding.txt')
#list_freq_sorted = output_list_freq_sorted(r'..\datasets\mimiciii_train_th0.txt')
list_x_axis = [ele for ele in range(1,len(list_freq_sorted)+1)]
plt.title('Distribution of labels for the MIMIC-III-shielding dataset')
plt.plot(list_x_axis,list_freq_sorted,'k+',label='MIMIC-III-shielding')
plt.xlabel('Index of ICD-9 code in MIMIC-III-shielding')
plt.ylabel('Number of occurrences')
#plt.legend(loc='upper right')
plt.xticks(np.arange(0, len(list_freq_sorted)+1, 5))
plt.savefig('MIMIC-III-shielding-label-dist.png',dpi=300)
```
